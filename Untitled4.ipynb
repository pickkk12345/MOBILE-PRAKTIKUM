**5. Visualisasikan dataset dengan salah satu grafik. Pilih grafik yang paling cocok untuk memvisualisasikan data tersebut.**

Kode ini fungsinya untuk membuat histogram distribusi jawaban responden:

Ambil hanya kolom pertanyaan Q1–Q55.

Gabungkan semua nilai jawaban responden (0–4).

Hitung berapa kali tiap skor muncul.

Plot histogram batang:

X (horizontal) = skor jawaban (0–4).

Y (vertikal) = jumlah total jawaban dari semua responden di semua pertanyaan.

Hasilnya grafik yang menunjukkan seberapa sering skor 0, 1, 2, 3, dan 4 dipilih.

```python
import matplotlib.pyplot as plt
import pandas as pd

questions = [col for col in df.columns if col.startswith("Q")]
data = df[questions]

value_counts = data.values.flatten()
counts = pd.Series(value_counts).value_counts().sort_index()

plt.bar(counts.index, counts.values, color='skyblue', edgecolor='black')

plt.xlabel("Skor Jawaban (1-4)")
plt.ylabel("Jumlah Responden (total dari semua Q1-Q55)")
plt.title("Distribusi Jawaban Responden (Gabungan Q1-Q55)")

plt.show()
```

**6. Tampilkan korelasi dataset menggunakan heatmap.**

Kode ini digunakan untuk menampilkan heatmap korelasi antar atribut dalam dataset:

df.corr() → menghitung korelasi antar semua kolom numerik (Q1–Q55 + label Divorce).

sns.heatmap(...) → membuat peta warna:

Merah → korelasi positif kuat.

Biru → korelasi negatif kuat.

Warna mendekati putih → korelasi lemah.

Sumbu X dan Y menampilkan semua atribut (pertanyaan dan label Divorce).

👉 Hasilnya: visualisasi yang memperlihatkan hubungan antar pertanyaan dan dengan label Divorce.

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = df.corr()

plt.figure(figsize=(15, 10))
sns.heatmap(corr, cmap="coolwarm", annot=False, cbar=True)

plt.title("Heatmap Korelasi Antar Pertanyaan dan Divorce", fontsize=16)
plt.show()
```
